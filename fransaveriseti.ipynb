```python
import pandas as pd
geography = pd.read_csv("2989_5652_bundle_archive/name_geographic_information.csv")
industry = pd.read_csv("2989_5652_bundle_archive/base_etablissement_par_tranche_effectif.csv")
salary = pd.read_csv("2989_5652_bundle_archive/net_salary_per_town_categories.csv")
population = pd.read_csv("2989_5652_bundle_archive/population.csv")
```

```python
geography.drop(['EU_circo', 'éloignement', 'numéro_circonscription', 'codes_postaux'], axis=1, inplace=True)
```

```python
geography.head()
```

```python
geography.rename(columns={'nom_région': 'region_name',
                          'code_région': 'region_insee',
                          'numéro_département': 'Department_INSEE',
                          'nom_département': 'Department_Name',
                          'préfecture':'Prefecture_Name',
                          'chef.lieu_région': 'region_capital',
                          'nom_commune': 'common_name'}, inplace=True)
```

```python
geography.head()
```

```python
geography.info()
```

```python
geography.dropna(subset = ["longitude", "latitude"], inplace=True)
```

```python
geography.info()
```

```python
geography["longitude"] = geography["longitude"].apply(lambda x: str(x).replace(',','.'))
mask = geography["longitude"] == '-'
geography.drop(geography[mask].index, inplace=True)
geography["longitude"] = geography["longitude"].astype(float)
```

```python
geography.info()
```

```python
geography.drop_duplicates(subset=["code_insee"], keep="first", inplace=True)
```

```python
geography.info()
```

```python
industry.head()
```

```python
industry.info()
```

```python
industry = industry[industry["CODGEO"].apply(lambda x: str(x).isdigit())]
```

```python
industry["CODGEO"] = industry["CODGEO"].astype(int)
```

```python
industry.head()
```

```python
salary.head()
```

```python
salary.drop(['LIBGEO'], axis=1, inplace=True)
```

```python
salary["CODGEO"] = salary["CODGEO"].astype(int)
```

```python
salary.rename(columns = {'SNHM14': 'mean_net_salary',
                        'SNHMF14': 'mean_net_salary_female',
                        'SNHMH14': 'mean_net_salary_male',
                        'SNHMC14': 'mnsh_executive',
                        'SNHMFC14': 'mnsh_executive_female',
                        'SNHMHC14': 'mnsh_executive_male',
                        'SNHMP14': 'mnsh_middlemanager',
                        'SNHMFP14': 'mnsh_middlemanager_female',
                        'SNHMHP14': 'mnsh_middlemanager_male',
                        'SNHME14': 'mnsh_employee',
                        'SNHMFE14': 'mnsh_employee_female',
                        'SNHMHE14': 'mnsh_employee_male',
                        'SNHMO14': 'mnsh_worker',
                        'SNHMFO14': 'mnsh_worker_female',
                        'SNHMHO14': 'mnsh_worker_male',
                        'SNHM1814': 'mnsh_year18_25',
                        'SNHMF1814': 'mnsh_year18_25_female',
                        'SNHMH1814': 'mnsh_year18_25_male',
                        'SNHM2614': 'mnsh_year26_50',
                        'SNHMF2614': 'mnsh_year26_50_female',
                        'SNHMH2614': 'mnsh_year26_50_male',
                        'SNHM5014': 'mnsh_year50over',
                        'SNHMF5014': 'mnsh_year50over_female',
                        'SNHMH5014': 'mnsh_year50over_male'}, inplace=True)
```
